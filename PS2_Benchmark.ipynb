# $\text{Benchmark}$

```python
import pandas as pd
import numpy as np

# =========================
# 1) Read & keep branded
# =========================
df = pd.read_csv("cigarettes.csv", low_memory=False, encoding="utf-8")

# Drop unlabeled brands (blank only, per your rule)
df["brand"] = df["brand"].astype("string").str.strip()
df = df.loc[df["brand"].notna() & (df["brand"] != "")].copy()
print(f"Kept rows with brand: {len(df):,}")

# Normalize keys
df["store"] = pd.to_numeric(df["store"], errors="coerce").astype("Int64")
df["week"]  = pd.to_numeric(df["week"],  errors="coerce").astype("Int64")
df["upc"]   = df["upc"].astype("string").str.replace(r"\D", "", regex=True)  # digits-only UPC

# Keep valid basics
df = df.loc[df["store"].notna() & df["week"].notna() & df["upc"].notna()].copy()

# =========================
# 2) Units in PACKS via carton flag (+ size override for 10CT/10PK/CARTON)
# =========================
def pick_col(cols, *cands):
    for c in cands:
        if c in cols: return c
    low = {c.lower(): c for c in cols}
    for c in cands:
        if c.lower() in low: return low[c.lower()]
    return None

carton_col = pick_col(df.columns, "carton", "Carton", "CARTON")

# Normalize size once for overrides
s = df["size"].astype("string").str.upper().str.replace(r"\s+", " ", regex=True).str.strip()
is_10ct   = s.str.contains(r"\b10\s*CT\b", regex=True, na=False)
is_10pk   = s.str.contains(r"\b10\s*PK\b", regex=True, na=False)
is_carton = s.str.contains(r"\bCARTON\b",  regex=True, na=False)
looks_10  = is_10ct | is_10pk | is_carton

if carton_col is None:
    df["packs_per_item"] = np.where(looks_10, 10.0, 1.0)
else:
    c = pd.to_numeric(df[carton_col], errors="coerce").fillna(0).astype(int)
    df["packs_per_item"] = np.where(c == 1, 10.0, 1.0)
    # Override when carton==0 but size indicates 10CT/10PK/CARTON
    override = (c != 1) & looks_10
    df.loc[override, "packs_per_item"] = 10.0

# Valid rows
df = df.loc[(df["qty"] > 0) & (df["packs_per_item"] > 0)].copy()

# =========================
# 3) Prices & quantities in PACKS (Dominick rule)
# =========================
# row_revenue (USD) = price * move / qty
# pack_sales (packs) = move * packs_per_item
df["row_revenue"] = df["price"] * df["move"] / df["qty"]
df["pack_sales"]  = df["move"]  * df["packs_per_item"]

# =========================
# 4) year52 (52-week blocks) and collapse to ONE ROW per PRODUCT per MARKET
#    market = store × year52
# =========================
base_week = int(df["week"].min())           # if weeks start at 1, (week-1)//52 + 1
df["year52"] = ((df["week"] - base_week) // 52 + 1).astype("Int64")

def first_nonnull(s):
    return s.dropna().iloc[0] if s.notna().any() else np.nan

# --- define characteristics to preserve ---
# dummies we keep if present (take max): extend this list if you have more
dummy_cols = [
    "menthol","dlx","special","supslim","slim","generic","single","carton","pack_kw",
    "HP","SP","F"  # include if your dataset has these as 0/1
]
dummy_cols = [c for c in dummy_cols if c in df.columns]

# continuous attributes (first non-null)
cont_cols = [c for c in ["tar_mean","nic_mean","co_mean"] if c in df.columns]

# categorical attributes (first non-null)
cat_cols = [c for c in ["brand","size","pack"] if c in df.columns]

# We will aggregate:
# - totals: total_packs, total_rev
# - dummies: max
# - continuous: first_nonnull
# - categorical: first_nonnull
agg_dict = {
    "pack_sales": ("pack_sales","sum"),
    "row_revenue": ("row_revenue","sum"),
}
for c in dummy_cols:
    agg_dict[c] = (c, "max")
for c in cont_cols:
    agg_dict[c] = (c, first_nonnull)
for c in cat_cols:
    agg_dict[c] = (c, first_nonnull)

prod_market = (
    df.groupby(["store","year52","upc"], as_index=False, observed=True)
      .agg(**agg_dict)
      .rename(columns={
          "pack_sales": "total_packs",
          "row_revenue":"total_rev"
      })
)

# quantity-weighted average price per pack
prod_market["avg_pack_price"] = prod_market["total_rev"] / prod_market["total_packs"]
prod_market.loc[~np.isfinite(prod_market["avg_pack_price"]), "avg_pack_price"] = np.nan

# carry packs_per_item at product-market level (first non-null within the group)
if "packs_per_item" in df.columns:
    ppi = (df.groupby(["store","year52","upc"], observed=True)["packs_per_item"]
             .first()
             .reset_index(name="packs_per_item"))
    prod_market = prod_market.merge(ppi, on=["store","year52","upc"], how="left", validate="one_to_one")

# =========================
# 5) Year market size (global per year): 1.5 × max store total in that year
# =========================
store_year_total = (
    prod_market.groupby(["store","year52"], observed=True)["total_packs"]
               .sum()
               .reset_index(name="store_year_total_packs")
)
year_max_store = (
    store_year_total.groupby("year52", observed=True)["store_year_total_packs"]
                    .max()
                    .reset_index(name="max_store_total_in_year")
)

year_max_store["market_size_year"] = 1.5 * year_max_store["max_store_total_in_year"]

prod_market = prod_market.merge(
    year_max_store[["year52","market_size_year"]],
    on="year52", how="left", validate="many_to_one"
)

# =========================
# 6) Product market share (logit-ready)
# =========================
den = prod_market["market_size_year"].to_numpy()
num = prod_market["total_packs"].to_numpy()
prod_market["prod_mkt_share"] = np.where(den > 0, num / den, 0.0).astype("float64")
prod_market["prod_mkt_share"] = prod_market["total_packs"] / prod_market["market_size_year"]

# =========================
# 7) Diagnostics
# =========================
sum_share_storeyear = (
    prod_market.groupby(["store","year52"], observed=True)["prod_mkt_share"]
               .sum()
)
print("Mean ∑ shares per (store,year52):", float(sum_share_storeyear.mean()))
print("Max  ∑ shares per (store,year52):", float(sum_share_storeyear.max()))

# Inspect columns we kept
print("\nColumns in prod_market:")
print(list(prod_market.columns))

prod_market
```

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from statsmodels.iolib.summary2 import summary_col

# ======================================
# 1. Prepare data
# ======================================

df = prod_market.copy()

# Ensure valid share
df = df.loc[(df["prod_mkt_share"] > 0) & (df["prod_mkt_share"] < 1)].copy()
df["outside_share"] = 1 - df.groupby(["store","year52"])["prod_mkt_share"].transform("sum")
df["log_s"] = np.log(df["prod_mkt_share"] / df["outside_share"])
df = df.rename(columns={'avg_pack_price': 'price'})

# Handle collinearity — drop 'Carton' (keep 'Single') or vice versa
drop_col = "Carton"
if drop_col in df.columns:
    df = df.drop(columns=[drop_col])

# Make sure categorical vars are strings
df["brand"]  = df["brand"].astype("category")
df["store"]  = df["store"].astype("category")
df["year52"] = df["year52"].astype("category")

# ======================================
# 2. Build regression formula
# ======================================

# Characteristic controls (adjust list as needed)
char_vars = ["menthol", "dlx", "special", "supslim", "slim", "tar_mean", "nic_mean", "co_mean"]

# Fixed effects: store and year
formula = (
    "log_s ~ price + C(brand) + "
    + " + ".join(char_vars)
    + " + C(year52)"
)

# ======================================
# 3. Estimate the model
# ======================================

model = smf.ols(formula=formula, data=df)
result = model.fit(cov_type="cluster", cov_kwds={"groups": df["store"]})  # cluster SE by store

print(result.summary())
```
