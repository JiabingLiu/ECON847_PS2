```python
import numpy as np 

dom = pd.read_csv("Characteristics/upccig_with_tnco_HARD.csv")
ftc = pd.read_csv("Characteristics/ftc_1994_1998_tnco_clean.csv")


ftc = ftc[['Brand Name', 'Description', 'Size', 'F', 'NF', 'HP', 'SP', 'Ultra_Lt',
       'Lt', 'FF', 'Menthol', 'Generic', 'Dlx', 'Special', 'SupSlim', 'Slim',
         'Tar_value', 'Nic_value',
        'CO_value', 'manufacturer_tested', 'value_brand']]

value_brands = [
    "Basic", "Doral", "USA Gold", "Viceroy", "Pyramid", "L&M", "Monarch",
    "Eagle", "Cambridge", "Bristol", "Alpine", "Bucks", "Players",
    "Old Gold", "Style", "Heritage", "Richland", "Montclair", "Malibu",
    "American Lights", "Bull Durham", "Century", "Sterling", "Magna"
]
super_value_brands = ["GPC", "Raleigh"]
private_label_names = ["Private Label", "Store Brand", "PRIVATE STOCK"]

# --------------------------------
# 2. Unify TNCO and brand info
# --------------------------------
ftc = ftc.rename(columns={
    "Brand Name": "brand",
    "Tar_value": "tar",
    "Nic_value": "nicotine",
    "CO_value": "co"
})
dom.columns = dom.columns.str.lower().str.replace(r"\s+", '_')
ftc.columns = ftc.columns.str.lower().str.replace(r"\s+", '_')


#ftc["brand"] = ftc["brand"].fillna("").str.strip().str.title()
#ftc["brand_upper"] = ftc["brand"].str.upper()

for c in ["tar", "nicotine", "co"]:
    ftc[c] = pd.to_numeric(ftc[c], errors="coerce")

# --------------------------------
# 3. Superslim / Ultra-slim tagging
# --------------------------------
desc_ftc = ftc["description"].fillna("").astype(str).str.upper()
ftc["supslim"] = np.where(ftc.description.str.contains("Sup-Slim",case=False, na=False), 1, 0)

############## Takes a single row from hard and select closest ftc
def match_characteristics(ftc_info, hard_info):
    if all(hard_info.pack_kw==1):
        if any(ftc_info.hp==1):
            ftc_info= ftc_info[ftc_info.hp==1]
    if all(hard_info.slim==1):
        if any(ftc_info.slim==1):
            ftc_info= ftc_info[ftc_info.slim==1]
    if all(hard_info.supslim==1):
        if any(ftc_info.supslim==1):
            ftc_info= ftc_info[ftc_info.sup_slim==1]
    size = list(hard_info[hard_info.descrip==d]["size"])[0]
    if any(ftc_info["size"]==size):
        ftc_info= ftc_info[ftc_info["size"]==size]
    if all(hard_info.special==1):
        if any(ftc_info.special==1):
            ftc_info= ftc_info[ftc_info.special==1]
    return ftc_info

############## Select possible ftc matches for each category

value_brands = [
    "Basic", "Doral", "USA Gold", "Viceroy", "Pyramid", "L&M", "Monarch",
    "Eagle", "Cambridge", "Bristol", "Alpine", "Bucks", "Players",
    "Old Gold", "Style", "Heritage", "Richland", "Montclair", "Malibu",
    "American Lights", "Bull Durham", "Century", "Sterling", "Magna"
]
super_value_brands = ["GPC", "Raleigh", "Nabisco", "american", "BEST VALUE"]
private_label_names = ["Private", "Store"]


def patter_list(l):
    return '|'.join(l)

ftc_private = ftc[ftc.description.str.contains("generic|"+patter_list(private_label_names),case=False, na=False)|ftc.brand.str.contains("generic|"+patter_list(private_label_names),case=False, na=False)]
ftc_value = ftc[ftc.description.str.contains("value|"+patter_list(value_brands),case=False, na=False)|ftc.value_brand==1|ftc.brand.str.contains("value|"+patter_list(value_brands),case=False, na=False)]
ftc_super_value = ftc[ftc.description.str.contains("value|"+patter_list(super_value_brands),case=False, na=False)|ftc.value_brand==1|ftc.brand.str.contains("value|"+patter_list(super_value_brands),case=False, na=False)]


############## Define generic categories
mask = (dom['descrip'].isin([ 'NAME GENRIC CIGS','PRICE VALUE KING CAR','SUB GENERIC SINGLE']))
dom.loc[mask, "generic_hardcoded"] = 1

mask = (dom['descrip'].isin([ "WINCHESTER 20'S"]))
dom.loc[mask, "cigarette"] = 0
dom.loc[mask, "cigar"] = 1

generic_cigs = dom[(dom['generic_hardcoded'] == 1) & (dom['cigarette'] == 1)].drop_duplicates()
generic_cigs = generic_cigs[[ 'descrip', 'size', 'special', 'supslim', 'slim',
         'pack_kw', 'tnco_key_level',
       'tar_mean', 'nic_mean', 'co_mean', 'cigarette',
       'value',  'flavored', 'menthol',
       'premium','super_value']].drop_duplicates()

private = generic_cigs[generic_cigs.value==0]
private = private[private.super_value==0]

value = generic_cigs[generic_cigs.value==1]
value = value[value.super_value==0]

super_value = generic_cigs[generic_cigs.super_value==1]
super_value = super_value[super_value.value==0]

generic_cigs = generic_cigs[[ 'descrip', 'size', 'special', 'supslim', 'slim',
         'pack_kw', 'tnco_key_level',
       'tar_mean', 'nic_mean', 'co_mean', 'cigarette',
       'value',  'flavored', 'menthol',
       'premium','super_value']].drop_duplicates()

################# Loop over rows to match with closest FTC record

import numpy as np
import pandas as pd

def match_mean(df, ftc_matches, d, fallback_ftc):
    """
    Write mean TNCO to rows with DESCRIP == d.
    Use column-wise fallback: if tar (or nic/co) is NaN from ftc_matches,
    fill just that column from fallback_ftc.
    """
    # ensure numeric
    for c in ["tar","nicotine","co"]:
        if c in ftc_matches:
            ftc_matches[c] = pd.to_numeric(ftc_matches[c], errors="coerce")
        if fallback_ftc is not None and c in fallback_ftc:
            fallback_ftc[c] = pd.to_numeric(fallback_ftc[c], errors="coerce")

    # primary means (may contain NaN)
    prim = ftc_matches[["tar","nicotine","co"]].mean(numeric_only=True)

    # fallback per column if needed
    if fallback_ftc is not None and not fallback_ftc.empty:
        fb = fallback_ftc[["tar","nicotine","co"]].mean(numeric_only=True)
    else:
        fb = pd.Series({"tar": np.nan, "nicotine": np.nan, "co": np.nan})

    tar_val = prim["tar"]        if np.isfinite(prim["tar"])        else fb["tar"]
    nic_val = prim["nicotine"]   if np.isfinite(prim["nicotine"])   else fb["nicotine"]
    co_val  = prim["co"]         if np.isfinite(prim["co"])         else fb["co"]

    # last resort: leave as NaN if both primary and fallback missing for that column

    mask = (df["descrip"] == d)
    df.loc[mask, "tar_mean"] = tar_val
    df.loc[mask, "nic_mean"] = nic_val
    df.loc[mask, "co_mean"]  = co_val
    return df


for d in set(private.descrip):
    ftc_matches = match_characteristics(ftc_private, private[private.descrip==d])
    fallback_ftc = ftc_private
    generic_cigs = match_mean(generic_cigs,ftc_matches,d,fallback_ftc)

for d in set(value.descrip):
    ftc_matches = match_characteristics(ftc_value, value[value.descrip==d])
    fallback_ftc = ftc_value
    generic_cigs = match_mean(generic_cigs,ftc_matches,d,fallback_ftc)

for d in set(super_value.descrip):
    ftc_matches = match_characteristics(ftc_super_value, super_value[super_value.descrip==d])
    fallback_ftc = ftc_super_value
    generic_cigs = match_mean(generic_cigs,ftc_matches,d,fallback_ftc)


generic_cigs
```
